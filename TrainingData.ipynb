```python
import cv2
import numpy as np
import os
from PIL import Image

#cv2.face la thu vien de training hinh anh
recognizer = cv2.face.LBPHFaceRecognizer_create()

path = 'dataSet'

def getImageWithId(path):
    imagePaths = [os.path.join(path, f) for f in os.listdir(path)] #truy cap vao tat ca duong dan anh 
    
    #print(imagePaths)
    
    faces = []
    IDs = []
    
    for imagePath in imagePaths:
        faceImg = Image.open(imagePath).convert('L') #L la grayscale, convert thang PIL image
        
        faceNp = np.array(faceImg, 'uint8') #chuyen anh thanh dang array de train (cac ma tran diem anh?)
        
        print(faceNp)

        #cat va lay theo ptu
        Id = int(imagePath.split('\\')[1].split('.')[1]) 
        
        faces.append(faceNp)
        IDs.append(Id)
        
        cv2.imshow('Training',faceNp)
        cv2.waitKey(10)
   
    return faces, IDs


faces, Ids = getImageWithId(path)

#train models
recognizer.train(faces, np.array(Ids))

if not os.path.exists('recognizer'): #neu ma chua co
    os.makedirs('recognizer')

recognizer.save('recognizer/trainingData.yml')

cv2.destroyAllWindows()
    
```
