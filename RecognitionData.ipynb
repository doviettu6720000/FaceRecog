```python
import cv2
import numpy as np
import os
import sqlite3
from PIL import Image
from datetime import date
from datetime import datetime

#load tv
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_alt.xml") #nhan dien mat
recognizer = cv2.face.LBPHFaceRecognizer_create() #thu vien train hinh anh

recognizer.read('C:\\Users\\ADMIN\Desktop\\Python\\FR\\Face Recognition with Webcam\\recognizer\\trainingData.yml')

now = datetime.now()

def getProfile(id):
    conn = sqlite3.connect('E:\\Tai lieu hoc tap\\DB SQLite\\data.db')
    query = "SELECT * FROM people WHERE ID= "+ str(id)
    cursor = conn.execute(query)
    
    profile = None
    
    for row in cursor:
        profile = row
    
    conn.close()
    return profile
    
cap = cv2.VideoCapture(0) #truy cap camera

fontface = cv2.FONT_HERSHEY_SIMPLEX

while(True):   
    ret, frame = cap.read()
    
    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    
    faces = face_cascade.detectMultiScale(gray) #truyen vao anh gray
    
    for (x , y, w , h) in faces: #ve hinh vuong
        cv2.rectangle(frame, (x,y), (x+w,y+h), (0,255,0), 2) #(anh; toa do x-y; toa do tinh tien; khung mau xanh; do day`)
        
        roi_gray = gray[y:y+h, x:x+w] #cat anh ra de kiem tra xem co trong db k; mau` xam' 
        
        id, confidence = recognizer.predict(roi_gray) #nhan dien khuon mat duoc cat ra; lay id va do chinh xac
        print(recognizer.predict(roi_gray))
        '''print(confidence)'''
        
        if confidence<60:
            profile = getProfile(id) #tra ve profile trong db           
            if(profile != None):
                cv2.putText(frame,""+str(profile[1]), (x+5,y+h+30),fontface,1,(0,255,0),2 )
                #(anh; id; ten; dong id ten hien o dau; font chu; ti le;mau xanh;do day)               
        else:
            cv2.putText(frame, "Unknow", (x+5,y+h+30),fontface,1,(0,0,255),2 )
    
    cv2.imshow('Training',frame)
    
    if(cv2.waitKey(1)==ord('q')):
        break
        
cap.release()
cv2.destroyAllWindows()
```
